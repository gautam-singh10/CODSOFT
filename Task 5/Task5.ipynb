```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

## Data collection and Processing

##### Dataset link -> https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

```python
cc_data=pd.read_csv('creditcard.csv')
```

```python
# checking missing values
cc_data.isnull().sum()
```

```python
legit=cc_data[cc_data.Class==0]
fraud=cc_data[cc_data.Class==1]
```

## Undersampling
### Building a sample dataset with similiar normal transactions and fraudulent transactions

### No. of fraudulent transactions -> 492

```python
legit_sample=legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Splitting data into targets

```python
X= new_dataset.drop(columns='Class',axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

### Splitting data into Training and Testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Model Training

### Logistic Regression

```python
model=LogisticRegression()
```

```python
# training model with the training data
model.fit(X_train,Y_train)
```

## Model Evaluation

### Accuracy score

```python
#accuracy score on training data 
X_train_predic=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_predic,Y_train)
```

```python
print("Accuracy on training data: ",training_data_accuracy)
```

```python
#accuracy score on test data 
X_test_predic=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_predic,Y_test)
```

```python
print("Accuracy on test data: ",test_data_accuracy)
```

```python
print(classification_report(X_test_predic,Y_test))
```
